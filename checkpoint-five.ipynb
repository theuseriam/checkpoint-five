## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
X = NAKES['November21']
y = NAKES['FY21_July-Nov_20']

#I wasn't able to create models with the other data sets. I am able to crete models with the NAICS data set
# This will look at the impact one month has on the annual total
#When I improve on iterations I will be able to drill this down into value in a row and total
```

```python
# Splitting the variables as training and testing in order to build the model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)
                                                    
                                                    

        
#creating a liner regression looking at the month of November and how it relates FY 2021  
#These training sets are split into a 7:3 ratio
```

```python
# I am taking a look at the train dataset
X_train
y_train
```

```python
# Importing Statsmodels.api library from Stamodel package
import statsmodels.api as sm
                                                    
                                                    

# Adding a constant for the sample month to get an intercept
X_train_sm = sm.add_constant(X_train)
                                                    
```

```python
# Fitting the resgression line using 'OLS' Ordinary least Square
lr = sm.OLS(y_train, X_train_sm).fit()

# Printing the parameters of the straight line
lr.params
```

```python
# Performing a summary to list out all the different parameters of the regression line fitted
lr.summary()

# coeffecient and p-value - high coeff, low p-value meaning low statistical significance - it's plotted a few blocks down
#R-Squared Value 0.981 based on the previous visualizations it doesn't seem likely that 98% is accurate
#F-Statistic - has a high p statistic so the model is not significatly significant f-3632, p 47
```

```python
# Take a look at the train dataset
X_train
y_train
```

## Error Terms

Finally, plot your error terms!

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, 4.439e+05 + 3.6031*X_train, 'r') # I'm using the constant and coefficient values from the summary to plot
plt.show()
```

```python

# Predicting y_value using training data of X
y_train_pred = lr.predict(X_train_sm)

# Creating residuals from the y_train data and predicted y_data
res = (y_train - y_train_pred)

#this is created based on the y value the cy total in my example
```

```python
fig = plt.figure()
sns.histplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()

#the data has a weibull distribution indicating moderate skewness and kurtosis
```
